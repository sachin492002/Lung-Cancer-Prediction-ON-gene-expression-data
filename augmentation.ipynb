***Augmentation by Weighted Mixed Observations***

```python
import numpy as np
def aug1(num_samples, df):
    original_df = np.array(df.T)  
    p,n = original_df.shape
    synth_df =[]
    for i in range(num_samples):
        random_weights = np.random.rand(n, 1)
        new_df = original_df.dot(random_weights)
        synth_df.append(new_df.flatten())
    synth_df =np.vstack(synth_df)
    
    
    synth_df=synth_df/(np.max(synth_df)/np.max(df))
    
    synth_df= synth_df.astype(int)
    synth_df = pd.DataFrame(synth_df, columns=df.columns)
    augmented_df = pd.concat([df, synth_df])
    
    return augmented_df
```

Positive data augmentation

```python
df = aug1(500,pos_data)
df.reset_index()
df['label']=1
df
```

NEgative data augmentation

```python
df2 = aug1(1000,neg_data)
df2.reset_index()
df2['label']=0
df2
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

class Generator(nn.Module):
    def __init__(self, latent_dim, num_genes):
        super(Generator, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(latent_dim, 250),
            nn.ReLU(),
            nn.Linear(250, 500),
            nn.ReLU(),
            nn.Linear(500, 1000),
            nn.ReLU(),
            nn.Linear(1000, num_genes),
            nn.Tanh()
        )

    def forward(self, x):
        return self.layers(x)

class Discriminator(nn.Module):
    def __init__(self, num_genes):
        super(Discriminator, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(num_genes, 1000),
            nn.LeakyReLU(0.2),
            nn.Linear(1000, 500),
            nn.LeakyReLU(0.2),
            nn.Linear(500, 250),
            nn.LeakyReLU(0.2),
            nn.Linear(250, 1)
        )

    def forward(self, x):
        return self.layers(x)


def train_gan(generator, discriminator, data_loader, num_epochs, lr,latent_dim):
    criterion = nn.BCEWithLogitsLoss()
    optimizer_g = optim.Adam(generator.parameters(), lr=lr)
    optimizer_d = optim.Adam(discriminator.parameters(), lr=lr)

    for epoch in range(num_epochs):
        for data in data_loader:
            batch_size = data.shape[0]
            real_labels = torch.ones(batch_size, 1)
            fake_labels = torch.zeros(batch_size, 1)

            # Train the discriminator
            optimizer_d.zero_grad()

            real_output = discriminator(data)
            real_loss = criterion(real_output, real_labels)
            real_loss.backward()

            noise = torch.randn(batch_size, latent_dim)
            fake_data = generator(noise)
            fake_output = discriminator(fake_data.detach())
            fake_loss = criterion(fake_output, fake_labels)
            fake_loss.backward()

            optimizer_d.step()

            # Train the generator
            optimizer_g.zero_grad()

            fake_output = discriminator(fake_data)
            gen_loss = criterion(fake_output, real_labels)
            gen_loss.backward()

            optimizer_g.step()

        # Print the loss for this epoch
        if epoch % 10 == 0:
            print(f"Epoch: {epoch+1}, Discriminator Loss: {real_loss+fake_loss}, Generator Loss: {gen_loss}")



def aug2(num_genes,data,no_samples):
    # Set hyperparameters
    latent_dim = 100
    
    num_epochs = 200
    lr = 0.0001
    batch_size = 32

    data_tensor = torch.Tensor(data)

    data_loader = DataLoader(data_tensor, batch_size=batch_size, shuffle=True)

    generator = Generator(latent_dim, num_genes)
    discriminator = Discriminator(num_genes)

    train_gan(generator, discriminator, data_loader, num_epochs, lr,latent_dim)


    noise = torch.randn(no_samples, latent_dim)
    aug_data = generator(noise).detach().numpy()
    

    aug_data_df = pd.DataFrame(aug_data)
    return  aug_data_df
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

class Generator(nn.Module):
    def __init__(self, latent_dim, num_genes):
        super(Generator, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(latent_dim, 250),
            nn.ReLU(),
            nn.Linear(250, 500),
            nn.ReLU(),
            nn.Linear(500, 1000),
            nn.ReLU(),
            nn.Linear(1000, num_genes),
            nn.Tanh()
        )

    def forward(self, x):
        return self.layers(x)

class Critic(nn.Module):
    def __init__(self, num_genes):
        super(Critic, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(num_genes, 1000),
            nn.LeakyReLU(0.2),
            nn.Linear(1000, 500),
            nn.LeakyReLU(0.2),
            nn.Linear(500, 250),
            nn.LeakyReLU(0.2),
            nn.Linear(250, 1)
        )

    def forward(self, x):
        return self.layers(x)

def gradient_penalty(critic, real_data, fake_data):
    batch_size = real_data.size(0)
    alpha = torch.rand(batch_size, 1)
    alpha = alpha.expand_as(real_data)
    interpolated = alpha * real_data + (1 - alpha) * fake_data
    interpolated.requires_grad_(True)

    critic_interpolated = critic(interpolated)
    gradients = torch.autograd.grad(
        outputs=critic_interpolated,
        inputs=interpolated,
        grad_outputs=torch.ones_like(critic_interpolated),
        create_graph=True,
        retain_graph=True,
        only_inputs=True,
    )[0]

    gradient_penalty = ((gradients.norm(2, dim=1) - 1) ** 2).mean()
    return gradient_penalty

def train_wgan(generator, critic, data_loader, num_epochs, lr, latent_dim, critic_iterations=5):
    optimizer_g = optim.Adam(generator.parameters(), lr=lr)
    optimizer_critic = optim.Adam(critic.parameters(), lr=lr)

    for epoch in range(num_epochs):
        for i, data in enumerate(data_loader):
            real_data = data

            for _ in range(critic_iterations):
                optimizer_critic.zero_grad()

                batch_size = real_data.shape[0]
                noise = torch.randn(batch_size, latent_dim)
                fake_data = generator(noise).detach()

                real_output = critic(real_data)
                fake_output = critic(fake_data)

                critic_loss = fake_output.mean() - real_output.mean() + 0.1 * gradient_penalty(critic, real_data, fake_data)
                critic_loss.backward()
                optimizer_critic.step()

            if i % critic_iterations == 0:
                optimizer_g.zero_grad()

                noise = torch.randn(batch_size, latent_dim)
                fake_data = generator(noise)

                fake_output = critic(fake_data)
                generator_loss = -fake_output.mean()
                generator_loss.backward()
                optimizer_g.step()

        
        if epoch % 10 == 0:
            print(f"Epoch: {epoch+1}, Critic Loss: {critic_loss.item()}, Generator Loss: {generator_loss.item()}")


def aug2_wgan(num_genes, data, no_samples):
    
    latent_dim = 100
    num_epochs = 200
    lr = 0.0001
    batch_size = 32
    critic_iterations = 5

    data_tensor = torch.Tensor(data)
    data_loader = DataLoader(data_tensor, batch_size=batch_size, shuffle=True)

    generator = Generator(latent_dim, num_genes)
    critic = Critic(num_genes)

    train_wgan(generator, critic, data_loader, num_epochs, lr, latent_dim, critic_iterations)

    noise = torch.randn(no_samples, latent_dim)
    aug_data = generator(noise).detach().numpy()

    aug_data_df = pd.DataFrame(aug_data)
    return aug_data_df
```
